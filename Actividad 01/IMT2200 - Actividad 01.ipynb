### Pregunta: ¿Cuál era la población total de Chile en 2017?

Para responderla podemos hacer un cálculo sencillo y sumar todos los datos de la columna `PERSONAS`:

```python
#Total de población
total_pop=dat['PERSONAS'].sum() # del data frame selecciona solo la columna "personas" y suma todo
print(f' La población total en 2017 fue: {total_pop} habitantes')
```

```python
# 
total_hombres=dat['HOMBRES'].sum() # Total población hombres
total_mujeres=dat['MUJERES'].sum() # Total población mujeres

total_filas=len(dat)
faltantes_homb=dat['HOMBRES'].isna().sum() # datos faltantes hombres
faltantes_muj=dat['MUJERES'].isna().sum() # datos faltantes mujeres

cobertura_muj=(total_filas-faltantes_muj)/total_filas*100
cobertura_homb=(total_filas-faltantes_homb)/total_filas*100

print(f'Se tiene un registro de {total_mujeres} mujeres durante el 2017')
print(f"Se tiene un registro de {total_hombres} hombres durante el 2017")

if faltantes_homb == faltantes_muj:
    print(f"Se tiene una cobertura del {cobertura_muj:,.0f}% tanto para hombres como para mujeres")
```




además de obtener el la población total de Chile en 2017 se realiza una medición de la cobertura de registros por género, definida como el porcentaje de filas del dataset que tienen datos válidos para cada género. En este caso, la cobertura es de 98% tanto para hombres como para mujeres ya que la cantidad de filas sin datos es la misma para ambos. 
Esto significa que, en los registros del Censo 2017, se cuenta con información de 8,949,887 mujeres y 8,569,735 hombres. El porcentaje faltante corresponde a registros en los que la cifra no está disponible, por lo que no es posible estimar la población total según género con exactitud. 



```python
etiquetas=["Hombres", "Mujeres"]
valores=[total_mujeres, total_hombres]
# Creamos gráfico de torta
plt.pie(valores, labels=etiquetas,autopct='%1.1f%%', startangle=90) # autopct me muestra los porcentajes en el gráfico
plt.title("HOMBRES V/S MUJERES AÑO 2017")
plt.show() # si no pongo parentésis me sale el tipo de gráfico
```

También podemos usar algunas funciones más avanzadas de `pandas`para analizar la distribución de personas por región y edad:

```python
#Distribución por edad y género
cols=['EDAD_0A5','EDAD_6A14','EDAD_15A64','EDAD_65YMAS']
for col in cols:
    dat[col]=dat[col].replace('*',0).astype(float)
    
pv=pd.pivot_table(dat, index='REGION', values=cols, aggfunc='sum', margins=False)
pv
```

Y podemos graficar la cantidad por región:

```python
fig = plt.figure(figsize=(20,4))
ax = fig.add_subplot(111)
ax.set_ylabel('Número de personas')
fig.suptitle('Distribución de población por Edad y Región (Chile, Censo 2017)')
pv.plot(kind='bar', ax =ax);
```

### Notemos que la población no está equitativamente distribuida por edades. 
Del gráfico se desprende que en 2017 la natalidad era baja, ya que la proporción de población entre 0 y 5 años es escasa. En contraste, la mayor parte de la población se concentra en el rango adulto, lo que sugiere una estructura demográfica con predominio de personas en edad laboral. 

Esta distribución podría implicar, a largo plazo, un envejecimiento progresivo de la población si la tendencia de baja natalidad se mantiene. 


## 4. Creación de un Mapa con información del Censo

A continuación usaremos librerías de geopandas para conectar un mapa de la Región Metropolitana con datos del Censo de 2017.

```python
#Leer datos censales
manz=gpd.read_file('R13/MANZANA_IND_C17.shp')
com=gpd.read_file('R13/COMUNA_C17.shp')
```

El siguiente código conecta el mapa de la Región Metropolitana con la población por manzana del Censo 2017. Edite el código para que la figura tenga su nombre en el Titulo

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.colors as colors

#Crear figura y ejes
fig=plt.figure(figsize=(13,13))
ax=fig.add_subplot(111)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# cantidad de personas por manzana
Z=manz['TOTAL_PERS']
manz.plot(ax=ax, column='TOTAL_PERS', cmap='Reds', legend=True, vmin=0, vmax=1000)

com_projected = com.to_crs(epsg=3857)
com_projected.boundary.plot(ax=ax, lw=0.2, color='k')

xlim=[-70.9,-70.4]
ylim=[-33.7,-33.2]
for x, y, label in zip(com_projected.geometry.centroid.x, com_projected.geometry.centroid.y, com_projected.NOM_COMUNA):
    if xlim[0]<x<xlim[1] and ylim[0]<y<ylim[1]:
        ax.text(x, y, label, fontsize = 10)
        
ax.set_xlim(xlim)
ax.set_ylim(ylim);

## Reemplace 'XXX' por su apellido
ax.set_title('RM - Población total por manzana - Censo 2017 - Ejecutado por Fernanda Le Roy',fontsize=10)

## Reemplace 'XXX' por su apellido en el nombre del archivo de salida 
fig.savefig('PoblacionRM_C2017_LeRoy.jpg')
```
